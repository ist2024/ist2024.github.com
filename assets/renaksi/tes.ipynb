```python
import os
import fitz  # PyMuPDF
from PIL import Image

# Path to the folder containing PDF files
pdf_folder = "D:\WAMP\www\ist2024.github.io\assets\renaksi"
output_folder = "D:\WAMP\www\ist2024.github.io\assets\renaksiimage"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through all PDF files in the folder
for filename in os.listdir(pdf_folder):
    if filename.endswith(".pdf"):
        # Open the PDF file
        pdf_path = os.path.join(pdf_folder, filename)
        pdf_document = fitz.open(pdf_path)
        
        # Extract the first page
        first_page = pdf_document.load_page(0)
        
        # Render page to image
        pix = first_page.get_pixmap()
        
        # Convert Pixmap to an Image object (Pillow)
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        
        # Save the image as PNG (or any other format)
        output_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.png")
        img.save(output_path)
        
        print(f"Saved first page of {filename} as an image.")

```
